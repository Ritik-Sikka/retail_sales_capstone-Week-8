```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("data/cleaned_data.csv")
```

```python
X = df[['Quantity','Price']]
y = df['Total_Sales']

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
from sklearn.cluster import KMeans

customer_data = df.groupby('Customer_ID')['Total_Sales'].sum().reset_index()

kmeans = KMeans(n_clusters=3, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(customer_data[['Total_Sales']])

customer_data.head()
```

```python
print(customer_data['Cluster'].value_counts())
print(customer_data.groupby('Cluster')['Total_Sales'].mean())
```
